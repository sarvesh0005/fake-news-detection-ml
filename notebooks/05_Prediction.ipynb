```python
import re
import string
import joblib

from pathlib import Path

import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("omw-1.4")
```

## Load Trained Model and Vectorizer

Load the saved TF-IDF vectorizer and the best-performing machine learning model.

```python
artifact_path = Path("../artifacts")

vectorizer = joblib.load(artifact_path / "tfidf_vectorizer.pkl")
model = joblib.load(artifact_path / "best_model.pkl")

print("Artifacts loaded successfully.")
```

## Text Preprocessing

The same preprocessing steps used during training are applied to the new input text to ensure consistency.

```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words("english"))


def preprocess_text(text):

    text = text.lower()

    text = re.sub(r"http\S+|www\S+", "", text)

    text = re.sub(r"<.*?>", "", text)

    text = text.translate(str.maketrans("", "", string.punctuation))

    text = re.sub(r"\d+", "", text)

    text = re.sub(r"\s+", " ", text).strip()

    words = text.split()

    words = [
        lemmatizer.lemmatize(word)
        for word in words
        if word not in stop_words
    ]

    return " ".join(words)
```

## Input News Article

Enter a news article to check whether it is predicted as Fake or Real.

```python
news = """
Donald Trump secretly signed an agreement with aliens to take over the White House next month.
"""
```

```python
clean_news = preprocess_text(news)

news_vector = vectorizer.transform([clean_news])

prediction = model.predict(news_vector)[0]

if prediction == 1:
    print("Prediction: Fake News")
else:
    print("Prediction: Real News")
```

```python
test_articles = [
    "NASA confirms water has been discovered on Mars after a new satellite mission.",
    "Donald Trump secretly signed an agreement with aliens to take over the White House next month.",
    "The Reserve Bank of India announced a change in interest rates today."
]

for article in test_articles:
    clean_article = preprocess_text(article)
    article_vector = vectorizer.transform([clean_article])

    prediction = model.predict(article_vector)[0]

    label = "Fake News" if prediction == 1 else "Real News"

    print("-" * 70)
    print("News:", article)
    print("Prediction:", label)
```
